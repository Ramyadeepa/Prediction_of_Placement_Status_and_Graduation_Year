```python
#function for predicting the year of graduation
def predict_year_of_graduation(academic_year, current_academic_year):
    current_year = datetime.date.today().year
    if 4 in {academic_year, current_academic_year}:
        return current_year
    elif 3 in {academic_year, current_academic_year}:
        return current_year + 1
    elif 2 in {academic_year, current_academic_year}:
        return current_year + 2
    elif 1 in {academic_year, current_academic_year}:
        return current_year + 3
    else:
        return 0
```

```python
df['Year of Graduation'] = pd.NA
#iterate over the rows of the dataframe
for index, row in df.iterrows():
    #get the column values for the current row
    academic_year = row['Academic Year']
    current_academic_year = row['What is your current academic year?']
    #call the function to predict the year of graduation
    predicted_year_of_graduation = predict_year_of_graduation(academic_year, current_academic_year)
    #save values in dataframe
    df.at[index, 'Year of Graduation'] = predicted_year_of_graduation
    #print the predicted values
    print(f"Index: {index}, Predicted Year of Graduation: {predicted_year_of_graduation}")
```
